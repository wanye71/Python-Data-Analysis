*loading a JSON file named 'allyears.json' into a pandas DataFrame named allyears. Loading JSON files into pandas DataFrames is a common way to work with structured data in Python.*

```python
allyears = pd.read_json('allyears.json')
```

```python
allyears_byyear = allyears.set_index(['sex', 'year']).sort_index()
```

```python
allyears_byyear.loc['M',2018].sort_values('number', ascending=False)
```

```python
allyears_byyear.loc['F',2018].sort_values('number', ascending=False).head(10).reset_index().name
```

```python
def getyear(sex, year):
    return (allyears_byyear.loc[sex, year]
           .sort_values('number', ascending=False)
           .head(10)
           .reset_index()
           .name)
```

```python
pd.DataFrame({year: getyear('M', year) for year in range(2010,2019)})
```

```python
pd.DataFrame({year: getyear('F', year) for year in range(2010,2019)})
```

```python
def plotname(sex, name):
    data = allyears.query('sex == @sex and name == @name')

    plt.plot(data.year, data.number, label=name)
    plt.axis(xmin=1880, xmax=2018)
```

```python
plt.figure(figsize=(12,5))

for name in getyear('F', 2018):
    plotname('F', name)

plt.legend()
```

```python
plt.figure(figsize=(12,5))

for name in getyear('M', 2018):
    plotname('M', name)

plt.legend()
```

```python
plt.figure(figsize=(12,5))

male_top_ten = ['Liam',
 'Noah',
 'Oliver',
 'Benjamin',
 'Elijah',
 'Lucas',
 'Mason',
 'Logan']

for name in male_top_ten:
    plotname('M', name)

plt.legend()
```

```python
alltime_females = allyears_byyear.loc['F'].groupby('name').sum().sort_values('number', ascending=False).head(5)
```

```python
alltime_females
```

```python
plt.figure(figsize=(12,5))

for name in alltime_females.index:
    plotname('F', name)

plt.legend()
```
