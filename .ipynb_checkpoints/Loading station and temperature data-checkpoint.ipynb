IV. FORMAT OF "ghcnd-stations.txt"

| Variable   | Columns  | Type         |
|------------|----------|--------------|
|  ID        |   1-11   |   Character  |
|  LATITUDE  |  13-20   |   Real       |
|  LONGITUDE |   22-30  |   Real       |
|  ELEVATION |   32-37  |   Real       |
|  STATE     |   39-40  |   Character  |
|  NAME      |   42-71  |   Character  |
|  GSN FLAG  |   73-75  |   Character  |
|  HCN/CRN FLAG   |   77-79  |   Character  |
|  WMO ID   |   81-85  |   Character  |

***Read data from a text file named stations.txt into a structured NumPy array. The genfromtxt() function allows you to specify the data types and formats for each column.***

***This code will read the data from stations.txt using the specified parameters and store it in the stations array. Make sure that the file stations.txt exists in the current directory and is formatted correctly according to the specified delimiter and column widths.***

```python
stations = np.genfromtxt('stations.txt', delimiter=[11,9,10,7,3,31,4,4,6],
                                         names=['id', 'latitude', 'longitude', 'elevation', 'state', 'name',
                                                'name', 'gsn', 'hcn', 'wmo'],
                                         dtype=['U11', 'd', 'd', 'd', 'U3', 'U31', 'U4', 'U4', 'U6'],
                                         autostrip=True)
```

```python
len(stations)
```

```python
stations
```

***This code will create a scatter plot where each point represents the longitude and latitude of a station. The markersize=1 argument sets the size of the marker dots.***

```python
plt.plot(stations['longitude'], stations['latitude'], '.', markersize=1)
```

***This line of code filters the stations array to select only the stations located in California (assuming 'CA' represents California in the 'state' column). The resulting array stations_ca will contain only the stations where the 'state' column equals 'CA'.***

***Now, stations_ca contains only the rows from stations where the 'state' column is equal to 'CA', effectively isolating the stations located in California. You can then use this filtered array for further analysis or visualization specific to California stations.***

```python
stations_ca = stations[stations['state'] == 'CA']
```

***This code will create a scatter plot where each point represents the longitude and latitude of a station in California. Adjust the marker size and other plot parameters as needed to customize the visualization.***




```python
plt.plot(stations_ca['longitude'], stations_ca['latitude'], '.', markersize=1)
```
