```python
import math
import collections
import itertools

import numpy as np
import pandas as pd
import matplotlib.pyplot as pp

%matplotlib inline
```

```python
words = sorted({line.strip().lower() for line in open('words.txt', 'r')})
```

```python
def signature(word):
    return ''.join(sorted(word))

words_by_sig = collections.defaultdict(set)

for word in words:
    words_by_sig[signature(word)].add(word)

anagrams_by_sig = {sig: wordset for sig, wordset in words_by_sig.items() if len(wordset) > 1}
```

```python
'wayne'[::-1]
```

```python
pairs = []

for wordset in anagrams_by_sig.values():
    for word1 in wordset:
        for word2 in wordset:
            # one way to check pairs only if they are sorted
            if word1 >= word2 and word1[::-1] == word2:
                pairs.append((word1, word2))
```

#### *Find pairs of words that are anagrams of each other but spelled in reverse*

```python
# initialize an empty list to store pairs of words that meet the criteria
pairs = [] 

# Iterates over the values of a dictionary
for wordset in anagrams_by_sig.values(): 
    #  iterates over all combinations of two words (word1 and word2)
    for word1, word2 in itertools.combinations(wordset, 2):
    # Checks if the reverse of word1 is equal to word2
        if word1[::-1] == word2:
            pairs.append((word1, word2))
```
